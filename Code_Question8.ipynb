```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('dark')

us_w_13=pd.read_csv('USOpen-women-2013.csv')
```

Q. Is there a variation among the playing style of the players as the tournament progresses and the opponent changes? Is this change of their playing style effective in them winning?

```python
us_w_13
```

```python
df=us_w_13[['Player 1', 'Player 2', 'Result', 'ACE.1', 'NPA.1', 'ACE.2', 'NPA.2']]
df=df.replace(np.nan, 0)
df
```

```python
unique_vals = pd.concat([df['Player 1'], df['Player 2']]).unique()

df1=pd.DataFrame(columns=['Player', 'Number of aces', 'Number of net shots attempted', 'Won/Lost'])

for i in range(len(unique_vals)):
    for index, row in df.iterrows():
        if unique_vals[i]==row['Player 1']:
            if row['Result']==0:
                s=pd.Series([unique_vals[i], row['ACE.1'], row['NPA.1'], 'L'], index=['Player', 'Number of aces', 'Number of net shots attempted', 'Won/Lost'])
                df1 = pd.concat([df1, s.to_frame().T], ignore_index=True)
            elif row['Result']==1:
                s=pd.Series([unique_vals[i], row['ACE.1'], row['NPA.1'], 'W'], index=['Player', 'Number of aces', 'Number of net shots attempted', 'Won/Lost'])
                df1 = pd.concat([df1, s.to_frame().T], ignore_index=True)
        elif unique_vals[i]==row['Player 2']:
            if row['Result']==0:
                s=pd.Series([unique_vals[i], row['ACE.1'], row['NPA.2'], 'W'], index=['Player', 'Number of aces', 'Number of net shots attempted', 'Won/Lost'])
                df1 = pd.concat([df1, s.to_frame().T], ignore_index=True)
            elif row['Result']==1:
                s=pd.Series([unique_vals[i], row['ACE.1'], row['NPA.2'], 'L'], index=['Player', 'Number of aces', 'Number of net shots attempted', 'Won/Lost'])
                df1 = pd.concat([df1, s.to_frame().T], ignore_index=True)
df1
```
