```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

sns.set_style('dark')

wim_m_13=pd.read_csv('Wimbledon-men-2013.csv')
```

Q. Which players completely dominated their opponent (won in straight sets) most number of times? What factor contributed most towards them dominating their opponent?

```python
df=pd.DataFrame(columns=wim_m_13.columns)

for index, row in wim_m_13.iterrows():
    if np.isnan(row['ST4.1']) and np.isnan(row['ST5.1']):
        df=pd.concat([df, row.to_frame().T], ignore_index=True)

df
```

Number of times each player won in straight sets

```python
arr=[]
for index, row in df.iterrows():
    if row['Result']==0:
        arr.append(row['Player2'])
    elif row['Result']==1:
        arr.append(row['Player1'])
        
arr=pd.Series(arr)
arr.value_counts()
```

```python
df1=pd.DataFrame(columns=wim_m_13.columns)

for index, row in df.iterrows():
    if row['Player1']=='A.Murray' or row['Player2']=='A.Murray':
        df1=pd.concat([df1, row.to_frame().T], ignore_index=True)
    elif row['Player1']=='N.Djokovic' or row['Player2']=='N.Djokovic':
        df1=pd.concat([df1, row.to_frame().T], ignore_index=True)
    elif row['Player1']=='J.Del Potro' or row['Player2']=='J.Del Potro':
        df1=pd.concat([df1, row.to_frame().T], ignore_index=True)
df1
```

```python
df2=pd.DataFrame(columns=['Player', 'First Serve Percentage', 'First Serve Won', 'Second Serve Percentage', 'Second Serve Won', 'Aces', 'Double Faults', 'Winners', 'Unforced Errors', 'Break Points Created', 'Break Points Won', 'Net Points Attempted', 'Net Points Won', 'Margin of win'])

for index, row in df1.iterrows():
    if row['Result']==0:
        s=pd.Series([row['Player2'], row['FSP.2'], row['FSW.2'], row['SSP.2'], row['SSW.2'], row['ACE.2'], row['DBF.2'], row['WNR.2'], row['UFE.2'], row['BPC.2'], row['BPW.2'], row['NPA.2'], row['NPW.2'], row['TPW.2']-row['TPW.1']], index=['Player', 'First Serve Percentage', 'First Serve Won', 'Second Serve Percentage', 'Second Serve Won', 'Aces', 'Double Faults', 'Winners', 'Unforced Errors', 'Break Points Created', 'Break Points Won', 'Net Points Attempted', 'Net Points Won', 'Margin of win'])
        df2=pd.concat([df2, s.to_frame().T], ignore_index=True)
    elif row['Result']==1:
        s=pd.Series([row['Player1'], row['FSP.1'], row['FSW.1'], row['SSP.1'], row['SSW.1'], row['ACE.1'], row['DBF.1'], row['WNR.1'], row['UFE.1'], row['BPC.1'], row['BPW.1'], row['NPA.1'], row['NPW.1'], row['TPW.1']-row['TPW.2']], index=['Player', 'First Serve Percentage', 'First Serve Won', 'Second Serve Percentage', 'Second Serve Won', 'Aces', 'Double Faults', 'Winners', 'Unforced Errors', 'Break Points Created', 'Break Points Won', 'Net Points Attempted', 'Net Points Won', 'Margin of win'])
        df2=pd.concat([df2, s.to_frame().T], ignore_index=True)

df2=df2.replace(np.nan, 0)
df2
```
